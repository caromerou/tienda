```python
import pandas as pd

# Leer los archivos subidos
df = pd.read_csv("/content/tienda_1 .csv")


# Mostrar uno como ejemplo
tienda1.head()
```

```python
df["Ingreso"] = df["Precio"]
```

```python
ingresos_totales = df["Ingreso"].sum()
print("Ingresos totales:", ingresos_totales)
```

```python
# Asumimos que cada fila es una unidad vendida
df["Ingreso"] = df["Precio"]  # Una venta = una unidad

# INGRESOS TOTALES
ingresos_totales = df["Ingreso"].sum()
print("Ingresos totales:", ingresos_totales)

# CATEGORÍAS MÁS VENDIDAS
categorias_mas_vendidas = df["Categoría del Producto"].value_counts()
print("\nCategorías más vendidas:")
print(categorias_mas_vendidas)

# RESEÑAS PROMEDIO POR PRODUCTO
reseñas = df.groupby("Producto")["Calificación"].mean().sort_values(ascending=False)
print("\nReseñas promedio por producto:")
print(reseñas)

# PRODUCTOS MÁS VENDIDOS
productos_mas_vendidos = df["Producto"].value_counts()
print("\nProductos más vendidos:")
print(productos_mas_vendidos)

# PRODUCTOS MENOS VENDIDOS
productos_menos_vendidos = productos_mas_vendidos.sort_values()
print("\nProductos menos vendidos:")
print(productos_menos_vendidos)

# COSTO DE ENVÍO PROMEDIO
costo_envio_promedio = df["Costo de envío"].mean()
print("\nCosto de envío promedio:", costo_envio_promedio)
```

```python
# CATEGORÍAS MÁS VENDIDAS
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
categorias_mas_vendidas.plot(kind='bar', color='skyblue')
plt.title("Categorías más vendidas")
plt.ylabel("Cantidad de ventas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# TOP 10 PRODUCTOS CON MEJORES RESEÑAS
plt.figure(figsize=(10,5))
reseñas.head(10).plot(kind='bar', color='orange')
plt.title("Top 10 productos mejor calificados")
plt.ylabel("Calificación promedio")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# PRODUCTOS MÁS VENDIDOS
plt.figure(figsize=(10,5))
productos_mas_vendidos.head(10).plot(kind='bar', color='green')
plt.title("Top 10 productos más vendidos")
plt.ylabel("Cantidad de ventas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# PRODUCTOS MENOS VENDIDOS
plt.figure(figsize=(10,5))
productos_menos_vendidos.head(10).plot(kind='bar', color='red')
plt.title("Top 10 productos menos vendidos")
plt.ylabel("Cantidad de ventas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
